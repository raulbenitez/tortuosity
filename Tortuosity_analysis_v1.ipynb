```
import pandas as pd
d = pd.read_csv('/content/Tortuosity_data_analysys_6_attributes_age.csv',sep=',',index_col=0)
d.head(20)
```

Remove observations with NaNs:

```
d.dropna(axis='rows',inplace=True)
d.head()
```

Primer fem un resum estadístic de les variables:

```
d.describe()
```

```
d.insert(7,'age_zone',0)
d.head()
```

```
d.loc[d['Age']<=60,'age_zone']=1
d.loc[(d['Age']>60) & (d['Age']<=70),'age_zone']=2
d.loc[(d['Age']>70) & (d['Age']<=80),'age_zone']=3
d.loc[d['Age']>80,'age_zone']=4
```

```
d1 = d.drop('Age',axis=1)
d1.head()
```

```
import seaborn as sns 
sns.pairplot(d1,hue='age_zone')
```

Representem només les tres variables de tortuositat:

```
import seaborn as sns 
sns.pairplot(d1,vars=['Aorta_t2','R-Iliac_t2','L_Iliac_t2'],kind="reg",hue='age_zone')
```

k-means

```
d2 = d.drop(['Aorta-l','R-Iliac-l','L-Iliac-l','age_zone','Age'],axis=1) # only tortuosity variables
d2.head()
```

```
from sklearn.cluster import KMeans
c_kmeans = KMeans(n_clusters=3).fit_predict(d2)
```

```
d2.insert(3,'kmeans',0)
```

```
d2['kmeans']=c_kmeans
d2.head()
```

```
sns.pairplot(d2,hue='kmeans')
```

```
from sklearn import mixture
import numpy as np
NMAX = 10
bic = []
for kG in np.arange(1,NMAX+1):
    gmm = mixture.GaussianMixture(n_components=kG,covariance_type='full').fit(d2)
    bic.append(gmm.bic(d2))
    print(bic)

idx_winner = np.argmin(bic)
print('...and the winner is GMM model G = {}'.format(idx_winner+1))
```

```
import matplotlib.pyplot as plt
plt.plot(bic)
```

```
gmm = mixture.GaussianMixture(n_components=3,covariance_type='full').fit(d2)

fig = plt.figure(figsize=(12,6))
c_gmm = gmm.fit_predict(d2)
```

```
d2.insert(4,'gmm',0)
```

```
d2['gmm']=c_gmm
sns.pairplot(d2,hue='gmm')
```

```
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(d2['Aorta_t2'], d2['R-Iliac_t2'], d2['L_Iliac_t2'], c=c_kmeans, s=60)
ax.view_init(30, 185)
plt.show()
```
